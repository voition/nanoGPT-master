```python
import jieba
import re 

def read_chinese_file_to_sentences(filepath):
    # 读取所有文本内容
    with open(filepath, 'r', encoding='utf-8') as file:
        text = file.read()
    
    # 使用正则表达式清洗文本：去除特殊符号、空白字符及换行符
    text = re.sub(r'[^\s\d\w\u4e00-\u9fff]', '', text) 
    text = re.sub(r'\s+', '', text)
    text = re.sub(r'\n', '', text)
    
    # 使用 jieba 进行中文分词
    words = jieba.cut(text)
    
    # 将分词结果转换为句子列表格式（模型输入要求）
    sentences = [list(words)]
    return sentences
```

```python
from gensim.models import Word2Vec

# 设置语料文件路径
file_path = 'sanguoyanyi.txt' 

# 获取处理后的分词列表 
sentences = read_chinese_file_to_sentences(file_path)

# 训练 Word2Vec 模型 
# 参数设置：词向量维度 vector_size=100，上下文窗口 window=5
model = Word2Vec(sentences, vector_size=100, window=5, min_count=8, workers=4)

# 保存模型 
model.save("sanguo_w2v.model")
```

```python
loaded_model = Word2Vec.load("sanguo_w2v.model") 
similar_words = loaded_model.wv.most_similar('刘备')

print("语义相似性（与“刘备”最接近的词）：")
for word, similarity in similar_words:
    print(f"{word}: {similarity:.4f}")
```

```python
# positive代表加项，negative代表减项
analogy_words = loaded_model.wv.most_similar(positive=['刘备', '张飞'], negative=['关羽'], topn=10)

print(f"\n类比推理（刘备 - 张飞 + 关羽 ）结果：")
for word, analogy in analogy_words:
    print(f"{word}: {analogy:.4f}")
```

```python
from gensim.models import Word2Vec

# 1. 加载训练好的三国演义模型 
loaded_model = Word2Vec.load("sanguo_w2v.model")

# 2. 设定需要对比的两个词语（例如人物名）
word1 = '刘备'
word2 = '曹操'

# 3. 计算两个词语的相似性 
# 注意：在 Word2Vec 模型中，需要通过 .wv 访问词向量进行计算 
if word1 in loaded_model.wv and word2 in loaded_model.wv:
    similarity = loaded_model.wv.similarity(word1, word2)
    
    # 4. 打印输出结果
    print(f"'{word1}' 与 '{word2}' 的语义相似度为: {similarity:.4f}")
else:
    print("模型词汇表中未找到输入的词语，请检查分词结果或降低 min_count 参数。")
```
